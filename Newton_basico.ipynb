```python
import numpy as np
```

```python
def f(x:list)->float:
  return np.float64((x[0]-x[1])**2+x[1]*(3-x[0]*x[0]))
```

```python
def d1f(x:list)->list:
  return [2*x[0]-2*x[1]-2*x[0]*x[1],-2*x[0]+2*x[1]-x[0]*x[0]+3]
```

```python
def d2f(x:list)->list:
  return np.matrix([[2-2*x[1],-2-2*x[0]],[-2-2*x[0],2]])
```

```python
x_0=[1,1]
l=1e-2
dk=np.matmul(np.linalg.inv(d2f(x_0)),-1*np.array(d1f(x_0)))
xk=x_0+[l*dk[0,0],l*dk[0,1]]
N=50
k=0
while np.linalg.norm(d1f(xk))>1e-3 and k<50:
  B=np.linalg.inv(d2f(xk))
  dk=np.matmul(B,np.array(d1f(xk)))
  xk=[xk[0]+l*dk[0,0],xk[1]+l*dk[0,1]]
  print(k,xk,f(xk))
  k+=1
```
