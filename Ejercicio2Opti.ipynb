```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
```

#Funcion a Optimizar
f(x,y)=xy

```python
def f(x):
    return x[0]*x[1]
```

```python
x = np.linspace(-1.5, 1.5, 100)
y = np.linspace(-1.5, 1.5, 100)
X, Y = np.meshgrid(x, y)
Z = f(np.vstack([X.ravel(), Y.ravel()])).reshape((100,100))
plt.figure(figsize=(10,8))
plt.contour(X, Y, Z, np.arange(-10,10, 0.05), cmap='jet');
plt.plot(x, np.sqrt(1-x**2), 'k:', linewidth=1)
plt.plot(x, -np.sqrt(1-x**2), 'k:', linewidth=1)
plt.xlabel('$x$'); plt.ylabel('$y$');
plt.title("Proyecci√≥n en el plano XY")
plt.axis([-1.5,1.5,-1.5,1.5])
```

```python
cons = ({'type': 'eq',
         'fun' : lambda x: np.array([x[0]**2+x[1]**2-1])})#,
         #'jac' : lambda x: np.array([1.0, 2.0*x[1]])})

bnds = ((-1.05, 1.05), (-1.05, 1.05))
```

```python
x0 = [np.sqrt(2)/2, np.sqrt(2)/2]
```

```python
ux = opt.minimize(f, x0, constraints=None)
ux
```

```python
cx = opt.minimize(f, x0, bounds=bnds, constraints=cons)
cx
```
